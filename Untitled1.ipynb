```python
# --- 1. IMPORT NECESSARY LIBRARIES ---
import requests
import pandas as pd
from datetime import datetime
import numpy as np

# --- 2. CONFIGURATION & SETUP ---
# Ensure your key is inside quotes.
# If you just created this key, it may take 2 hours to activate.
API_KEY = "d58827572da52467a12f0cb7f1ce3e8c"

# Focus districts for AES monitoring in Bihar
DISTRICTS = [
    "Muzaffarpur", "Vaishali", "Sitamarhi",
    "East Champaran", "Patna", "Gaya", "Saran"
]

# --- 3. DATA FETCHER FUNCTION ---
def fetch_bihar_weather(city_list, key):
    collected_data = []
    print(f"ðŸ”„ Starting live data fetch for {len(city_list)} districts...")

    for city in city_list:
        try:
            url = f"http://api.openweathermap.org/data/2.5/weather?q={city},IN&appid={key}&units=metric"
            response = requests.get(url)
            data = response.json()

            if response.status_code == 200:
                info = {
                    "District": city,
                    "Temperature": data['main']['temp'],
                    "Humidity": data['main']['humidity'],
                    "Condition": data['weather'][0]['main'],
                    "Timestamp": datetime.now().strftime("%Y-%m-%d %H:%M:%S")
                }
                collected_data.append(info)
            elif response.status_code == 401:
                print(f"âŒ API Key Error: Your key is likely still activating. Please wait 2 hours.")
                return pd.DataFrame() # Stop early if key is invalid
            else:
                print(f"âš ï¸ {city}: {data.get('message', 'Unknown error')}")
        except Exception as e:
            print(f"ðŸ’¥ Critical error for {city}: {e}")

    return pd.DataFrame(collected_data)

# --- 4. EXECUTION ---
live_weather_df = fetch_bihar_weather(DISTRICTS, API_KEY)

if not live_weather_df.empty:
    # Feature Engineering: Litchi Season (May/June)
    current_month = datetime.now().month
    live_weather_df['Litchi_Season'] = 1 if current_month in [5, 6] else 0

    print("\nâœ… Success! Live Bihar Data Ingested:")
    print(live_weather_df)
else:
    print("\nðŸ›‘ No data was fetched. Verify your API key status at openweathermap.org.")
```

```python
from sklearn.linear_model import LogisticRegression
import joblib  # To save the "brain" of our model

# --- 1. PREPARING THE TRAINING DATA ---
# We create a historical pattern: High Temp + High Humidity = High Risk (1)
np.random.seed(42)
train_size = 1000
X_train_data = {
    'Temperature': np.random.normal(35, 5, train_size),
    'Humidity': np.random.normal(65, 15, train_size),
    'Litchi_Season': np.random.choice([0, 1], train_size)
}
train_df = pd.DataFrame(X_train_data)

# Logic: Risk is 1 if Temp > 38 and Humidity > 70
train_df['Outcome'] = ((train_df['Temperature'] > 38) & (train_df['Humidity'] > 70)).astype(int)

# --- 2. TRAINING THE MODEL (Based on your high-performing logic) ---
# Your study showed Logistic Regression had the best AUC at 82.37%
X = train_df[['Temperature', 'Humidity', 'Litchi_Season']]
y = train_df['Outcome']

model = LogisticRegression()
model.fit(X, y)
print("âœ… Model Training Complete: Logistic Regression Brain is ready.")

# --- 3. PREDICTING RISK FOR LIVE BIHAR DISTRICTS ---
if not live_weather_df.empty:
    # We select the same features from our Live Feed
    live_features = live_weather_df[['Temperature', 'Humidity', 'Litchi_Season']]

    # Get the Probability of an outbreak (Class 1)
    # Just like you used predict_proba for AUC in your report
    risk_probabilities = model.predict_proba(live_features)[:, 1]

    # Add the Risk Score to our main table
    live_weather_df['Risk_Score (%)'] = (risk_probabilities * 100).round(2)

    # Add an Actionable Alert Status
    def get_status(score):
        if score > 70: return "ðŸ”´ HIGH ALERT (Deploy ASHA)"
        elif score > 40: return "ðŸŸ¡ MONITORING (Distribute ORS)"
        return "ðŸŸ¢ STABLE"

    live_weather_df['Action_Status'] = live_weather_df['Risk_Score (%)'].apply(get_status)

    print("\n--- BIHAR GOVT EARLY WARNING DASHBOARD ---")
    print(live_weather_df[['District', 'Temperature', 'Humidity', 'Risk_Score (%)', 'Action_Status']])
else:
    print("ðŸ›‘ No live data found. Please run Step 1 again first.")
```

```python
import folium

# --- 1. COORDINATES FOR BIHAR DISTRICTS ---
# We define the latitude and longitude for the districts we are monitoring
district_coords = {
    "Muzaffarpur": [26.1209, 85.3647],
    "Vaishali": [25.9934, 85.1110],
    "Sitamarhi": [26.5925, 85.4851],
    "East Champaran": [26.6469, 84.9089],
    "Patna": [25.5941, 85.1376],
    "Gaya": [24.7914, 85.0002],
    "Saran": [25.8560, 84.8564]
}

# --- 2. INITIALIZE THE MAP ---
# Centered in the middle of North Bihar
bihar_map = folium.Map(location=[25.8, 85.2], zoom_start=8, tiles='cartodbpositron')

# --- 3. PLOT LIVE RISK DATA ON THE MAP ---
if not live_weather_df.empty:
    for index, row in live_weather_df.iterrows():
        district_name = row['District']
        if district_name in district_coords:
            coords = district_coords[district_name]
            risk = row['Risk_Score (%)']
            status = row['Action_Status']

            # Choose color based on Status
            if "HIGH" in status: color = 'red'
            elif "MONITOR" in status: color = 'orange'
            else: color = 'green'

            # Create a popup with full medical/environmental details
            popup_text = f"""
                <b>District:</b> {district_name}<br>
                <b>Temperature:</b> {row['Temperature']}Â°C<br>
                <b>Humidity:</b> {row['Humidity']}%<br>
                <b>AES Risk Score:</b> {risk}%<br>
                <b>Status:</b> {status}
            """

            # Add a circle marker for each district
            folium.CircleMarker(
                location=coords,
                radius=15,
                popup=folium.Popup(popup_text, max_width=300),
                color=color,
                fill=True,
                fill_color=color,
                fill_opacity=0.6
            ).add_to(bihar_map)

    # Save the map as an HTML file (Great for LinkedIn!)
    bihar_map.save('Bihar_AES_Risk_Map.html')
    print("âœ… Interactive Map Generated! Check the 'Bihar_AES_Risk_Map.html' file in Colab.")
else:
    print("ðŸ›‘ No data found to map. Run Step 1 and 2 first.")

# Display the map in Colab
bihar_map
```
